```python
import pandas as pd

Outlook = ['Rainy', 'Sunny', 'Overcast', 'Overcast', 'Sunny', 'Rainy', 'Sunny',
          'Overcast', 'Rainy', 'Sunny', 'Sunny', 'Rainy', 'Overcast', 'Overcast']

Play = ['Yes', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'Yes',
          'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes']

dataset = {
   'Outlook': Outlook,
   'Play':Play,
        }

arr = [i+1 for i in range(len(dataset['Outlook']))]
df = pd.DataFrame(dataset, index = arr)
df
```

```python
df.to_csv('Weather Conditions.csv', header=False, index=False) 
```

```python
w = df.groupby('Outlook').size()
w
```

```python
w.sum() 
```

```python
x = df.where(df['Play']=='Yes').groupby('Outlook').size()
x
```

```python
y = df.groupby('Play').size()
y
```

```python
z = df.where(df['Play']=='No').groupby('Outlook').size()
z
```

# Applying Bayes'theorem:

# p(y|s) = ( p(s|y) * p(y) ) / p(s) 



```python
psy = x[2]/x.sum()
psy
```

```python
ps = w[2]/w.sum()
ps
```

```python
py = y[1]/y.sum()
py
```

```python
pys = psy*py/ps
pys
```

```python
psn = z[1]/z.sum()
psn
```

```python
pn = y[0]/y.sum() 
pn
```

```python
pns = psn*pn/ps
pns
```

```python
if pys > pns:
    print('On a Sunny day, Player can play the game') 
else:
    print('On a Sunny day, Player can NOT... play the game') 
```
