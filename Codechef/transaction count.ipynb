```python
def grouptransaction(str):
    str = str.split()
    str2 = []

    for i in str:
        if i not in str2:
            str2.append(i)

    for i in range(0, len(str2)):
        print('Frequency of', str2[i], 'is :', str.count(str2[i]))	

def main():
    transactions = ['notebook', 'notebook', 'mouse', 'keyboard', 'mouse']
    str = ', '.join(transactions)
    grouptransaction(str)

if __name__=="__main__":
    main() 
```

```python
def grouptransaction(transactions):
    import pandas as pd
    counts = pd.Series(transactions).value_counts()
    final = []

    for i,j in counts.items():
        final.append(f'{i} {j}')
    return final
    
if __name__=="__main__":
    transactions = ['notebook', 'notebook', 'mouse', 'keyboard', 'mouse'] 
    final = grouptransaction(transactions)
    print(final) 
```
